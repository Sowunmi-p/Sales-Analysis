```python
import seaborn as sns
#sns.set()
sns.set(style="whitegrid")


import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
S_df = pd.read_csv('Downloads/sales2.csv')
```

```python
S_df.head()
```

```python
S_df.duplicated().sum()
```

```python
display(S_df.drop_duplicates(keep=False,inplace=True))
```

```python
S_df.info()
```

```python
S_df.describe().transpose()
```

```python
S_df.isnull().sum()
```

```python
S_df['Customer_Gender'].replace('M', 'Male', inplace = True)
S_df['Customer_Gender'].replace('F','Female',inplace = True)
```

```python
S_df.head()
```

```python
#Age distribution of Cusomers
Dis = S_df.Age_Group.value_counts()
Dis
```

```python
plt.figure(figsize=(12,6))
plt.title('Age Distribution')
plt.pie(Dis,labels=Dis.index, autopct='%1.1f%%', startangle=180, shadow=False)
```

```python
plt. figure (figsize =(12,6))
plt.title("Age_group")
plt.xlabel('Customer_Age')
plt.ylabel('frequency')
plt. hist(S_df.Customer_Age, bins= 10, color ='blue', edgecolor='white');
```

```python
#most profitable year
most_p = S_df.groupby('Year')['Profit'].sum().sort_values(ascending=False).reset_index()
most_p
```

```python
plt.figure(figsize=(12,6))
plt.title('Most_profitable')
sns.barplot(x=most_p.index,y=most_p,palette='Set2')
```

```python
#most protitable month
most_pm = S_df.groupby(['Month','Year'])['Profit'].sum().sort_values(ascending=False)
most_pm
```

```python
S_df.groupby(['Month','Year'])['Profit'].sum().sort_values(ascending=False).plot(kind='bar', figsize=(20,10))
```

```python
#Gender with the most orders
Gender_order = S_df.groupby('Customer_Gender')['Order_Quantity'].sum().sort_values(ascending=False)
Gender_order
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender_orders')
plt.pie(Gender_order,labels=Gender_order.index, autopct='%1.1f%%', startangle=180, shadow=False)
```

```python
#country with the highest revenue
highestR = S_df.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
highestR
```

```python
plt.figure(figsize=(12,6))
plt.title('highestR')
plt.pie(highestR,labels=highestR.index, autopct='%1.1f%%', startangle=180, shadow=False)
```

```python
#State with the highest revenue
highestRS = S_df.groupby('State')['Revenue'].sum().sort_values(ascending=False).head(10)
highestRS
```

```python
plt.figure(figsize=(12,6))
plt.title('highestRS')
plt.pie(highestRS,labels=highestRS.index, autopct='%1.1f%%', startangle=180, shadow=False)
```

```python
plt.figure(figsize=(20,10))
plt.title('Highest_State_Revenue')
sns.barplot(x=highestRS.index,y=highestRS,palette='Set2')
```

```python
#most profit category
most_pmp = S_df.groupby(['Product_Category','Sub_Category'])['Profit'].sum().sort_values(ascending=False)
most_pmp
```

```python
#created another dataframe for year 2012 only
customer_2012 = S_df.iloc[:,3]==2012
customer_20_40 = np.logical_and(S_df.iloc[:,4]>=20, S_df.iloc[:,4]<=40)
c_df = S_df[customer_2012 & customer_20_40]
c_df
```

```python
#the most occuring Age for customers in year 2012
c_df.Customer_Age.value_counts()
```

```python
#getting the product with the most sales
most_pmpp = c_df.groupby(['Product_Category','Sub_Category'])['Profit'].sum().sort_values(ascending=False)
most_pmpp
```

```python
plt.figure(figsize=(12,6))
plt.title('h')
plt.pie(most_pmpp,labels=most_pmpp.index, autopct='%1.1f%%', startangle=180, shadow=False)
```

```python
#created another dataframe for year 2013
onlycustomer_2013 = S_df.iloc[:,3]==2013
customer_20= S_df.iloc[:,4]<=20
Cost_df = S_df[customer_2013 & customer_20]
Cost_df
```

```python
#the most occuring Age for customers in year 2013
Cost_df.Customer_Age.value_counts()
```

```python
#the country with most cost
cost = Cost_df.groupby('Country')['Cost'].sum().sort_values(ascending=False)
cost
```

```python
plt.figure(figsize=(12,6))
plt.title('Country_mostcost')
plt.pie(cost,labels=cost.index, autopct='%1.1f%%', startangle=180, shadow=False)
```
